```python
# Dengue Analysis - Rodriguez Markdave
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("dengue.csv")
```

#  helps identify which regions have the highest share of reported dengue cases.

```python
region_cases = df.groupby("Region")["Dengue_Cases"].sum().sort_values(ascending=False)
plt.figure(figsize=(6,6))
plt.pie(region_cases, labels=region_cases.index, autopct="%1.1f%%")
plt.title("Dengue Cases by Region")
plt.show()

```

# insights 1 : highlights the region with the most dengue cases, showing where the outbreak is most concentrated.


```python
top_region = region_cases.head(1)
plt.bar(top_region.index, top_region.values)
plt.title("Insight 1: Top Region with Highest Dengue Cases")
plt.xlabel("Region")
plt.ylabel("Cases")
plt.show()
```

# INSIGHT 2 : the region least affected by dengue, providing contrast to the most affected areas.

```python
lowest_region = region_cases.tail(1)
plt.bar(lowest_region.index, lowest_region.values)
plt.title("Insight 2: Region with Lowest Dengue Cases")
plt.xlabel("Region")
plt.ylabel("Cases")
plt.show()
```

# INSIGHT 3 : This bar chart compares the top three regions, emphasizing which areas contribute most to dengue cases.

```python
top3 = region_cases.head(3)
plt.bar(top3.index, top3.values)
plt.title("Insight 3: Top 3 Regions Comparison")
plt.xlabel("Region")
plt.ylabel("Cases")
plt.show()
```

# Insight 4: This pie chart shows the share of dengue cases coming from the top three regions, giving a clear proportional view.

```python
plt.pie(top3, labels=top3.index, autopct="%1.1f%%")
plt.title("Insight 4: Percentage of Top 3 Regions")
plt.show()
```

# Insight 5:This bar chart visualizes all regions to show how dengue cases are distributed across the country.

```python
plt.bar(region_cases.index, region_cases.values)
plt.title("Insight 5: Overall Dengue Distribution by Region")
plt.xlabel("Region")
plt.ylabel("Cases")
plt.xticks(rotation=90)
plt.show()
```



# This heatmap displays correlations between dengue-related variables (like cases, deaths, and population). It helps determine how strongly each variable is connected to one another.

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title(" Correlation Heatmap")
plt.show()
```

# Insight 6 :This heatmap shows the correlation between dengue cases and deaths.A high correlation indicates that more cases often lead to more fatalities.

```python
plt.figure(figsize=(4,3))
sns.heatmap(df[["Dengue_Cases", "Dengue_Deaths"]].corr(), annot=True)
plt.title("Insight 6: Dengue Cases vs Deaths")
plt.show()
```

# Insight 7 :This heatmap examines how dengue cases vary over different years, showing possible trends or spikes.

```python
plt.figure(figsize=(4,3))
sns.heatmap(df[["Year", "Dengue_Cases"]].corr(), annot=True)
plt.title("Insight 7: Year vs Dengue Cases")
plt.show()
```

# Insight 8:This visualization explores how deaths due to dengue relate to the year, which may reflect improvements in healthcare or changes in outbreak intensity.

```python
plt.figure(figsize=(4,3))
sns.heatmap(df[["Year", "Dengue_Deaths"]].corr(), annot=True)
plt.title("Insight 3: Year vs Deaths")
plt.show()
```

# insight 9 : Displays how dengue deaths vary each month, possibly linked to rainfall or temperature changes.

```python
avg_deaths = df.groupby("Region")["Dengue_Deaths"].mean().sort_values(ascending=False)
plt.figure(figsize=(6,4))
sns.barplot(x=avg_deaths.values, y=avg_deaths.index)
plt.title("Insight 9: Average Dengue Deaths per Region")
plt.xlabel("Average Deaths")
plt.ylabel("Region")
plt.show()
```

# insights 10 : The scatter plot shows that higher dengue cases often lead to more deaths, confirming the positive relationship observed in the heatmap.

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df["Dengue_Cases"], y=df["Dengue_Deaths"])
plt.title("Insight 10: Relation Between Dengue Cases and Deaths")
plt.xlabel("Dengue Cases")
plt.ylabel("Dengue Deaths")
plt.show()

```
