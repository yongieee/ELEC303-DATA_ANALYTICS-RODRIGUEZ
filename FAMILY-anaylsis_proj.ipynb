# Family Income and Expenditure Analysis: Descriptive and Predictive Modeling



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset and clean column names
df = pd.read_csv("Family Income and Expenditure.csv")
df.columns = df.columns.str.strip()

# Set plotting style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python

income_source_counts = df['Main Source of Income'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(income_source_counts, labels=income_source_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Main Household Income Sources', fontsize=16)
plt.tight_layout()
plt.show()
```

# Insights 1:
# Dominant Wage Reliance: Wage/Salaries is the dominant source of income for the majority of households, representing over $49\%$ of the total, indicating a primary dependency on the formal employment sector.

# Insights 2:
# Significant Entrepreneurial Activity: Income from Entrepreneurial Activities makes up a significant one-third ($24.8\%$) of the total, describing a large informal or self-employed sector in the economy..

# Insights 3:
# Economic Focus: The small size of the Other Sources category ($26.1\%$) describes the limited role of non-traditional income streams like remittances or social assistance for most households in the sample

```python

region_income = df.groupby('Region')['Total Household Income'].mean().sort_values(ascending=False).reset_index()
plt.figure(figsize=(14, 7))
sns.barplot(x='Region', y='Total Household Income', data=region_income, palette='viridis')
plt.title(' Average Total Household Income by Region (Inequality)', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Average Total Household Income (PHP)', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.ticklabel_format(style='plain', axis='y')
plt.tight_layout()
plt.show()
```

# Insights 4:
# Regional Disparity Identified: The chart clearly shows significant differences in Average Total Household Income across the regions, diagnosing unequal economic development across the country.
# Insights 5:
# Economic Hubs Located: The highest bars indicate the major economic hubs (e.g., NCR), while the lowest bars pinpoint regions that may be economically lagging and require targeted attention.

```python

income_by_demographics = df.groupby(['Household Head Sex', 'Household Head Marital Status'])['Total Household Income'].mean().reset_index()
plt.figure(figsize=(12, 7))
sns.barplot(x='Household Head Marital Status', y='Total Household Income', hue='Household Head Sex', data=income_by_demographics, palette='Set1')
plt.title('3. Average Total Household Income by Head Sex and Marital Status', fontsize=16)
plt.xlabel('Household Head Marital Status', fontsize=12)
plt.ylabel('Average Total Household Income (PHP)', fontsize=12)
plt.xticks(rotation=0, ha='center', fontsize=10)
plt.ticklabel_format(style='plain', axis='y')
plt.legend(title='Household Head Sex')
plt.tight_layout()
plt.show()
```

# Insights 6 :
# Marital Status as a Predictor: For both male and female household heads, being Married is consistently associated with the highest average household income, predicting that the combined earning power of a couple is the greatest driver of household wealth.
# Insights 7 :
# Gender Income Gap Predictor: Within most marital statuses, Male household heads show a higher average income compared to Female household heads, suggesting that gender, combined with marital status, is a strong predictor of household income level.
# Insights 8 :
# Key Predictive Segment: The highest bar (likely Married Males) represents the most financially robust segment, making this combination the strongest demographic predictor of high household income in the dataset.

```python

edu_exp_by_source = df.groupby('Main Source of Income')['Education Expenditure'].mean().sort_values(ascending=False).reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Main Source of Income', y='Education Expenditure', data=edu_exp_by_source, palette='cividis')
plt.title('4. Avg. Education Expenditure by Main Source of Income (Investment)', fontsize=16)
plt.xlabel('Main Source of Income', fontsize=12)
plt.ylabel('Average Education Expenditure (PHP)', fontsize=12)
plt.xticks(rotation=0, ha='center', fontsize=10)
plt.ticklabel_format(style='plain', axis='y')
plt.tight_layout()
plt.show()
```

# Insights 9:
Wage Income Drives Investment: Households whose main income source is Wage/Salaries spend the highest average amount on education, explaining that stable, formal employment encourages the highest investment in future human capital.
# Insights 10:
Trade-off in Entrepreneurship: Entrepreneurial households, while having high incomes, allocate less to education than wage earners, possibly explaining a trade-off where capital is prioritized for business reinvestment over high educational spending.
