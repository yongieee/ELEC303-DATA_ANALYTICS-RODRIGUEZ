# Family Income and Expenditure Analysis: Descriptive and Predictive Modeling



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset and clean column names
df = pd.read_csv("Family Income and Expenditure.csv")
df.columns = df.columns.str.strip()

# Set plotting style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python

income_source_counts = df['Main Source of Income'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(income_source_counts, labels=income_source_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('1. DESCRIPTIVE: Distribution of Main Household Income Sources', fontsize=16)
plt.tight_layout()
plt.show()
```

# Insights 1:
# Dominant Wage Reliance: Wage/Salaries is the dominant source of income for the majority of households, representing over $60\%$ of the total, indicating a primary dependency on the formal employment sector.

```python

region_income = df.groupby('Region')['Total Household Income'].mean().sort_values(ascending=False).reset_index()
plt.figure(figsize=(14, 7))
sns.barplot(x='Region', y='Total Household Income', data=region_income, palette='viridis')
plt.title('2. DIAGNOSTIC: Average Total Household Income by Region (Inequality)', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Average Total Household Income (PHP)', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.ticklabel_format(style='plain', axis='y')
plt.tight_layout()
plt.show()
```

```python
# ==============================================================================
# 3. PREDICTIVE: Average Total Household Income by Head Sex and Marital Status (Grouped Bar Chart)
# ==============================================================================
# Calculate the mean income grouped by Sex and Marital Status
income_by_demographics = df.groupby(['Household Head Sex', 'Household Head Marital Status'])['Total Household Income'].mean().reset_index()
plt.figure(figsize=(12, 7))
sns.barplot(x='Household Head Marital Status', y='Total Household Income', hue='Household Head Sex', data=income_by_demographics, palette='Set1')
plt.title('3. PREDICTIVE: Average Total Household Income by Head Sex and Marital Status', fontsize=16)
plt.xlabel('Household Head Marital Status', fontsize=12)
plt.ylabel('Average Total Household Income (PHP)', fontsize=12)
plt.xticks(rotation=0, ha='center', fontsize=10)
plt.ticklabel_format(style='plain', axis='y')
plt.legend(title='Household Head Sex')
plt.tight_layout()
plt.show()
```

```python
# ==============================================================================
# 4. EXPLANATORY: Avg. Education Expenditure by Main Source of Income (Bar Chart)
# ==============================================================================
# Calculate the mean education expenditure by income source
edu_exp_by_source = df.groupby('Main Source of Income')['Education Expenditure'].mean().sort_values(ascending=False).reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Main Source of Income', y='Education Expenditure', data=edu_exp_by_source, palette='cividis')
plt.title('4. EXPLANATORY: Avg. Education Expenditure by Main Source of Income (Investment)', fontsize=16)
plt.xlabel('Main Source of Income', fontsize=12)
plt.ylabel('Average Education Expenditure (PHP)', fontsize=12)
plt.xticks(rotation=0, ha='center', fontsize=10)
plt.ticklabel_format(style='plain', axis='y')
plt.tight_layout()
plt.show()
```
